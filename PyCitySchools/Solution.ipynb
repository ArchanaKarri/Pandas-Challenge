```python
# Dependencies and Setup
import pandas as pd
import pandas as geek

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
#Counting the number of schools from the school_data DataFrame
no_schools=school_data['school_name'].count()
#Counting the number of students from the student_data DataFrame
no_students=student_data['student_name'].count()
#Calculating the total budget from the school_data DataFrame
tot_budget=school_data['budget'].sum()
#Calculating the average maths score from the school_data_complete DataFrame using mean 
avg_maths=school_data_complete['maths_score'].mean()
#Calculating the average reading score from the school_data_complete DataFrame using mean 
avg_reading=school_data_complete['reading_score'].mean()
#Calculating the number of studnets who scored more than 50 in maths  from the school_data_complete DataFrame using mean 
more_maths= school_data_complete[school_data_complete['maths_score'] >=50]['student_name'].count()
#Calculating the percentage of students who scored more than 50 in maths  from the school_data_complete DataFrame using mean 
per_maths=more_maths/no_students*100
#Calculating the number of studnets who scored more than 50 in reading from the school_data_complete DataFrame using mean 

more_reading= school_data_complete[school_data_complete['reading_score'] >=50]['student_name'].count()
#Calculating the percentage of students who scored more than 50 in reading  from the school_data_complete DataFrame using mean 

per_reading=more_reading/no_students*100
#Calculating the number of students who scored more than 50 in maths and reading from the school_data_complete DataFrame using mean 

more_mathsreading=school_data_complete[
    (school_data_complete['reading_score'] >=50)
    & (school_data_complete['maths_score'] >=50)]['student_name'].count()
#Calculating the percentageof students who scored more than 50 in maths and reading from the school_data_complete DataFrame using mean 
over_pass=more_mathsreading/no_students*100
#Creating a dictionary using all the values calculated above
dict = {'Total Schools':[no_schools],
        'Total Students': [no_students],
        'Total Budget':[tot_budget],
        'Average Maths Score':[avg_maths],
        'Average Reading Score':[avg_reading],
        '% Passing Maths':[per_maths],
        '% Passing Reading':[per_reading],
        '% Overall Passing':[over_pass]
       }
#formatting the budget in $ and 2 points after decimal format
lga_df=pd.DataFrame(dict)
lga_df['Total Budget']=lga_df['Total Budget'].map("${:,.2f}".format)
lga_df # LGA DataFrame
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Filtering the data using school name and retrieving the type of school data
schooltype=school_data_complete.groupby(["school_name"])["type"].unique()
schooltype.str.get(0)
#Calculating the number of students in each school
schoolstudents=school_data_complete.groupby('school_name')['type'].count()
#Calculating the budget for  each school
schoolbudget=school_data.groupby(['school_name'])['budget'].sum()
#Calculating the budget allocated per students in each school
perstudent=schoolbudget/schoolstudents
#Calculating average maths score in each school
avg_mathscore=school_data_complete.groupby(['school_name'])['maths_score'].mean()
#Calculating average reading score in each school
avg_readingscore=school_data_complete.groupby(['school_name'])['reading_score'].mean()
##Calculating who got more than 50  maths score in each school
maths_abovelimit = school_data_complete[school_data_complete['maths_score'] >= 50]
##Counting who got more than 50  maths score in each school
maths_count=maths_abovelimit.groupby('school_name')['student_name'].count()
#counting the percentage of students who scored more than 50 in maths for each school
maths_count=(maths_count/schoolstudents)*100
#collecting the data of students who scored more than 50 in reading for each school
reading_abovelimit = school_data_complete[school_data_complete['reading_score'] >= 50]
#counting the number of students who scored more than 50 in reading for each school
reading_count=reading_abovelimit.groupby('school_name')['student_name'].count()
#counting the percentage of students who scored more than 50 in maths for each school
reading_count=(reading_count/schoolstudents)*100
#collecting the data of students who scored more than 50 in maths and reading for each school
mathsreading_abvlimit=school_data_complete[(school_data_complete['maths_score'] >= 50)& (school_data_complete['reading_score'] >= 50)]
#counting the number  of students who scored more than 50 in maths and reading for each school
mathsreading_abvlimit=mathsreading_abvlimit.groupby('school_name')['student_name'].count()
#counting the percentage of students who scored more than 50 in maths and reading for each school
per_pass=(mathsreading_abvlimit/schoolstudents)*100
#Creating the dictionary to hold the data for each school
school_summarydict={"School Type":schooltype.str.get(0),
                "Total Students":schoolstudents,
                "Total School Budget":schoolbudget.map("${:,.2f}".format),
                "Per Student Budget":perstudent.map("${:,.2f}".format),
                "Average Maths Score":avg_mathscore,
                "Average Reading Score":avg_readingscore,
                "% Passing Maths":maths_count,
                "% Passing Reading":reading_count,
                "% Overall Passing":per_pass
                   }
#Creating the DataFrame using the data for each school
schoolsummary_df=pd.DataFrame(school_summarydict)
schoolsummary_df.head()
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Calculating the maths mean data for Year 9 for each school in Series
df_year9=school_data_complete.loc[school_data_complete['year']==9]

new_df_year9maths=df_year9.groupby(['school_name'])['maths_score'].mean()

result9_maths=pd.Series(new_df_year9maths)

# Calculating the maths mean data for Year 10 for each school in Series
df_year10=school_data_complete.loc[school_data_complete['year']==10]

new_df_year10maths=df_year10.groupby(['school_name'])['maths_score'].mean()
result10_maths=pd.Series(new_df_year10maths)

# Calculating the maths mean data for Year 11 for each schoolin Series
df_year11=school_data_complete.loc[school_data_complete['year']==11]

new_df_year11maths=df_year11.groupby(['school_name'])['maths_score'].mean()

result11_maths=pd.Series(new_df_year11maths)
# Calculating the maths mean data for Year 12 for each school in Series
df_year12=school_data_complete.loc[school_data_complete['year']==12]

new_df_year12maths=df_year12.groupby(['school_name'])['maths_score'].mean()
result12_maths=pd.Series(new_df_year12maths)
# Combining the all series into DataFrame for Year 9 to 12 for Maths
new_maths=pd.concat([result9_maths,result10_maths,result11_maths,result12_maths],axis=1)
new_maths.columns.values[0] = 'Year 9'
new_maths.columns.values[1] = 'Year 10'
new_maths.columns.values[2] = 'Year 11'
new_maths.columns.values[3] = 'Year 12'
#DataFrame with consolidated math score for each school based on year level
new_maths
```
